```python
dataset[['Midterm_Score', 'Final_Score', 'Assignments_Avg', 'Quizzes_Avg', 'Projects_Score']].mean()
```

```python
print("Распределение оценок:")
dataset['Grade'].value_counts().sort_index()
```

```python
print("Топ-5 студентов с наивысшим итоговым баллом:")
dataset.nlargest(5, 'Total_Score')[['First_Name', 'Last_Name','Gender' , 'Total_Score', 'Grade','Sleep_Hours_per_Night']]
```

```python
print("Среднее количество часов сна по факультетам:")
dataset.groupby('Department')['Sleep_Hours_per_Night'].mean()
```

```python
print("Средний итоговый балл студентов с доступом к интернету и без:")
dataset.groupby('Internet_Access_at_Home')['Total_Score'].mean()
```

```python
print("Столбцы, содержащие 'Score':")
dataset.filter(like='Score').head()
```

```python
print("Строки с индексами от 10 до 15:")
dataset.filter(items=range(10, 15), axis=0)
```

```python
print("Студенты с итоговым баллом выше 90:")
dataset[dataset['Total_Score'] > 90]['First_Name']
```

```python
print("Студенты с итоговым баллом выше 85 и уровнем стресса ниже 5:")
dataset[(dataset['Stress_Level (1-10)'] > 7) & (dataset['Sleep_Hours_per_Night'] < 6)][['First_Name','Age','Grade']]
```

```python
print("Студенты с высоким уровнем стресса и недостатком сна:")
dataset[(dataset['Stress_Level (1-10)'] > 7) & (dataset['Sleep_Hours_per_Night'] < 6)][['First_Name','Age','Grade']]
```
